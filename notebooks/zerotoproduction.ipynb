## There is intrinsic complexity in data

## Discovery, how dow e know we are making a something people want?
### From outcomes to solutions
- Product outcome: Change in human behaviour driving business impact
- Opportunitiy: Unmet needs or desires that satisfy the outcome
- Solution: How are we going to solve it

### Solutions clearly linked to outcomes
These levels show how our day-t-day solution work impacts product outcomes (and ultimate business goals)

## Discovery-delivery cycles Deciding what to build + building that what
4 risks\
1. Feasibility risk
Can we build it?
2. Value risk
3. Usability risk
4. Viability risk

### Some assumption and risks
Solution - Verify income through sharing back account transactions\
- feasibility risk assumption - we can accurately identify income from bank transaction data and distinguish from other payments.
- Value risk assumption - Consumers are willing to share their bank transactions to secure a tenancy
- Viability risk assumption - We treat income from disability payments or social security in a non-discriminatory manner

## Tackle the biggest risk first, most important assumptions with the weakest evidence
### Discovery as a Data Scientist
Discovery is a team sport\
Be part of the discovery process from the start\
Data science projects are dominated by feasibility (DS-led) and value (Product-led) risks\
Common discovery techniques\
- Qualitative: customers interviews
- Quantitative: data analysis

### Managing scope creep during discovery
Data discovery - testing assumptions through analysis and model building aim to (in)validate teh assumption, and no more\

Intrinsic uncertainty in teh data makes work estimation during planning difficult\
Decrease risk of scope creep and rabbit holes through regular check ins\
Data will raise questions during discovery <- Outcome = vital context\

### Deliver to support validatied assumptions
Another scope creep check! Scope to validated assumptions only\
True for code, data, model complexity\
Develop defensively against unvalidated assumption

### Deliver good code
Agile is greate here end-to-end value delivery fits nicely\
Production quality - use Engineering design patterns, coding standards, tests...\

### Building and keeping trust
Get agreement early and often
- Ask stakeholders to trade off priorities in the problem space
- Agree outcomes and any linked OKRs - contract between team and rest of organisastion
- State what you are doing - and what you are NOT doing - before you start

Report progress towards outcomes regularly: sprint reviews, project updates...\

Huge value in building a first solution\
- People don't always engage without something tangible
- Gives confidence to stakeholders, keeps them invested
