### Filter methods
Considered ranking methods since they are appplied to feature to rank their importance before classification or regression.\
Advantages\
- Scale well to large datasets while maintaining computational efficiency
- Easy to implement.

Disadvantages\
- Lack the ability to consider

### WRapper methods
Use the predictor as a black box and evaluate the performance of the ML models on subset of features.\
Advantages\
- Performs well in scenarios where feature interactions are important.
- Easy to implement.

Disadvantages\
- Their tendency to overfit on smaller datasets
- High computational cost limits their applicability to big dataset

### Embedded methods
Incorporate feature selection directly into the model training process.\
Advantages\
- Embedded methods have the advantage of being integrated directly into the learning algo, which allows them to account for feature interactions naturally.
- Better computational efficiency as compared to Wrapper Methods, they execute faster.

Disadvantages\
- Can be sensitive to choice of regularisation

## Motivation
From feature to shapley values\
- Shapley valuyes describe the contribution of each input featre to the value of a function.
- The sum of shapley values across all features MUST equal the value of the function on that data point
- For tree-based model, there is a fast algo for calculating them, implemented in shap package

- To control for overfitting, lets look at shapley values oin the validation dataset
- As Shapley values MUST add up to the final function value, let's do linear regression of the target on the SHapley Values
    - For a perfect fit, all coefficients will be 1.0
    - Negative coefficient means this feature is pull ing the function away
